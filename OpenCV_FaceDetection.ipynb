```
#import opencv
import cv2
#import matplotlib for graphs/plots
import matplotlib.pyplot as plt
```

```
#path of input image
image_path = '/content/drive/My Drive/Colab Notebooks/people2.jpg'
#path face detection cascade file which contains features of face
cascade_file ='/content/drive/My Drive/Colab Notebooks/haarcascades/haarcascade_frontalface_default.xml'
```

```
#function to mark rectangle on faces on people present in the image
def show_faces(image_path):
    #create a Cascade classifier using face detection cascade file.
    face_cascade = cv2.CascadeClassifier(cascade_file)
    
    #read a image
    img  = cv2.imread(image_path, 1)

    #convert image from one color face to another(here from RGB->GRAY)
    gray_img = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
    
    #detects faces and returns list containing co-ordinates of faces
    faces = face_cascade.detectMultiScale(  gray_img, scaleFactor=1.05, minNeighbors=5)
    
    #create rectangles covering faces on image
    for x,y,w,h in faces:
        img = cv2.rectangle(img, (x,y), (x+w, y+h), (0,255,0), 2)
    
    #resize the output image.
    #resized = cv2.resize(img, (int(img.shape[1]/2), int(img.shape[0]/2)))
    return img
```

```
#function to process input image and gives call to above function.
def face_detect(image_path):
    #reading a original image
    original_image = cv2.imread(image_path, 1)
    #processed output image
    out_image = show_faces(image_path)
    #function call to draw a graph of input & output image
    plot_graph(original_image, out_image)

#plot a graph of original and processed image.
def plot_graph(original_img, out_img):
    #create a figure with one row and two colomns.
    fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(20, 8))

    ax1.imshow(original_img)
    ax1.set_title('Original Image')
 
    ax2.imshow(out_img)
    ax2.set_title('Output Image')
```

```
image_path = '/content/drive/My Drive/Colab Notebooks/people2.jpg'
#FINAL CALL
face_detect(image_path)
```
